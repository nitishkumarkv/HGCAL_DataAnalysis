```python
def cnt(lst, end_lay):
    return np.array([lst.count(i) for i in range(1,end_lay+1)])

def apply(df):
    return cnt(df.rechit_layer, 40)

def lay_energy(num_hit_lay, rechit_energy, end_lay):
    return np.array([sum(rechit_energy[sum(num_hit_lay[0:i]) : (sum(num_hit_lay[0:i])+num_hit_lay[i])]) for i in range(end_lay)])

def app(df):
    return lay_energy(df.num_hits_layer, df.rechit_energy, 40)

```

```python
pion_data = pd.read_pickle("combined_data_positron_shuffled.pkl")
pion_data[:5]
```

```python
pion_data["num_hits_layer"]= pion_data.rechit_layer.map(lambda x : cnt(list(x),40))
```

```python
pion_data[:5]
```

```python
pion_data["rechit_energy_layer"] = pion_data.apply(app, axis=1)
pion_data["rechit_energy_Ecal"] = pion_data.rechit_energy_layer.map(lambda x : sum(x[:28]))
pion_data["rechit_energy_Hcal"] = pion_data.rechit_energy_layer.map(lambda x : sum(x[28:]))
pion_data["rechit_energy_Hcal_by_Ecal"] = pion_data.rechit_energy_Hcal / pion_data.rechit_energy_Ecal
pion_data["std_x"] = pion_data.rechit_x.map(lambda x : np.std(x))
pion_data["std_y"] = pion_data.rechit_y.map(lambda x : np.std(x))
pion_data[:5]
```
